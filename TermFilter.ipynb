```python
# chose columns to load
keep_columns = ["created_at", "full_text", "user.screen_name", "retweet_count", "favorite_count", "entities.hashtags", "in_reply_to_screen_name", "entities.user_mentions"]  # TODO: fill in with column names you want to keep!
```

```python
allmeps = pd.read_csv('meps_in_out.csv')
allmeps = allmeps[allmeps["TwitterHandle"].notna()].reset_index()
```

```python
# keep only current
current_meps = allmeps[allmeps["LeaveDate"].isna()].reset_index()
keep_cols = ["Name", "Country", "Party", "TwitterHandle", "JoinDate"]
current_meps = current_meps[keep_cols]
```

```python
# load the data only reading relevant columns
for f in keepcsvs[100:]:
    tweets = pd.read_csv(f)
    if tweets.empty:
        print(f, "does not tweet")
    else:
        try:
            tweets = tweets[keep_columns]
            try:
                tweets["created_at"] = pd.to_datetime(tweets["created_at"])
                name = f.split('/')[1].split('.')[0] # file name to Twitter handle
                i = current_meps.index[current_meps.TwitterHandle == name] # find in the df of MEPs
                start_date = current_meps.iloc[i, current_meps.columns.get_loc("JoinDate")].item() # because it returns an object
                dataRel = tweets.loc[tweets.created_at >= start_date]
                
                # save a new file
                name = 'termtime/' + name + '.csv'
                dataRel.to_csv(name)
            except:
                print(tweets.loc[tweets["user.screen_name"]])
        except:
            print(f + ": there is a problem with this user")
```
